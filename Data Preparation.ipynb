```python
def find_outliers_IQR(df):
   q1=df.quantile(0.25)
   q3=df.quantile(0.75)
   IQR=q3-q1
   outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]
   return outliers
```

```python
Mathes_outliers = find_outliers_IQR(df['Mathes'])
print('number of outliers:'+ str(len(Mathes_outliers)))
print('max outlier value:' + str(Mathes_outliers.max()))
print('min outlier value:'+ str(Mathes_outliers.min()))
Mathes_outliers
```

### Part A 

```python
df1=pd.read_csv('employee.csv')
df1.head()
```

```python
bins = [18, 30, 40, 50,60]
labels = ['19-30', '31-40', '41-50','51-60']
df1['age'] = pd.to_numeric(df1['age'], errors='coerce')
df1['Age_Category'] = pd.cut(df1['age'], bins=bins, labels=labels)
```

```python
df1.head()
```

### Part B

```python
label_encoder = LabelEncoder()
df1['gender'] = label_encoder.fit_transform(df1['gender'])
df1.head()
```

```python
# One-Hot Encoding using Pandas
df1 = pd.get_dummies(df1, columns=['marital_status'], dtype='int')
# Displaying the first few rows of the updated DataFrame
df1.head()
```

### Part C

```python
#Month and Year
df1['doj'] = pd.to_datetime(df1['doj'], format="%d-%m-%Y")
df1['month']=df1['doj'].dt.month
df1['year']=df1['doj'].dt.year
```

```python
df1.head()
```

```python
from datetime import datetime
current_date = datetime.now()
# Step 12: Calculate number of years passed 
df1['Passed_years'] = df1['doj'].apply(lambda x: current_date.year - x.year if pd.notnull(x) else np.nan)
df1.head()
```

```python
#Number of months passed
df1['Passed_months'] = df1['doj'].apply(lambda x: (current_date.year - x.year) * 12 + current_date.month - x.month if pd.notnull(x) else np.nan)
df1.head()
```

```python
df1['day_name']=df1['doj'].dt.day
df1.head()
```
