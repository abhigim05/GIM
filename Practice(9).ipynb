```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('sales.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
# Total number of missing values
df.isnull().sum().sum()
```

```python
df.isna().sum()
```

```python
# Define the garbage values to be replaced
garbage_values = ['&&&','***',"%%%",'$$$']

# Replace garbage values with NaN
df.replace(garbage_values,inplace=True)
df.head()
```

```python
df['bucket']=pd.cut(pd.to_numeric(df['Unit price']),3,labels=['low price','mid-price','high price'])
df.head()
```

```python
#applying min-max normalisation
df['normalized_profit'] = (df['Profit'] - df['Profit'].min()) / (df['Profit'].max() - df['Profit'].min())
#check the data contents
df.head()
```

```python
df['Z_Total']=(df['Total']-df['Total'].mean())/df['Total'].std()
df.head()
```

```python
mode=df['Quantity'].mode()[0]
df['Quantity'].fillna(mode,inplace=True)
df
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
category=['Product line']
encoder = OneHotEncoder(sparse_output=False)
one_hot_encoded = encoder.fit_transform(df[category])
one_hot_df = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(category))

df1=pd.concat([df,one_hot_df],axis=1)
df1.head()
```

```python
# Import label encoder
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()
```

```python
df['Customer type'] = label_encoder.fit_transform(df['Customer type'])
df.head(10)
```

```python
mean=df['gross income'].mean()
df['gross income'].fillna(mean,inplace=True)
df
```
